https://www.acmicpc.net/problem/15654

중복 x, 정렬

```python
n, m = map(int, input().split())
lst = list(map(int, input().split()))
visited = [False] * (n + 1)
answer = []
```

```python
lst.sort()
```

```python
def series(lth, n, m):
    if lth == m:
        print(' '.join(map(str, answer)))
        
    else:
        for i in range(1, n+1):
            if not visited[i]:
                visited[i] = True
                answer.append(lst[i-1])
                series(lth + 1, n, m)
                answer.pop()
                visited[i] = False
```

```python
series(0,n,m)
```

```python
def series(lth, n, m):
    if lth == m:
        print(' '.join(map(str, answer)))
        
    else:

        if not answer:
            k = 1
        else:
            a = answer[-1]
            k = lst.index(a) + 1

        for i in range(k, n+1):
            answer.append(lst[i-1])
            series(lth + 1, n, m)
            answer.pop()
```
